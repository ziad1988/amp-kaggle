```python
targets = ['updrs_1', 'updrs_2', 'updrs_3', 'updrs_4'] 
mean_dict = {i: train_clinical[i].mean() for i in targets}
mean_dict
```

```python
import amp_pd_peptide
env = amp_pd_peptide.make_env()
iter_test = env.iter_test()
```

```python
count = 0 
for (test, test_peptides, test_proteins, sample_submission) in iter_test:
    
    targets = sample_submission.prediction_id.str.split('_').str[2:4].apply(lambda x: '_'.join(x))
    if count == 0: 
        print (targets)
    print (10*'---sample submission ---')
    print (sample_submission.head())
    print (10*'------')
    
    
    sample_submission['rating'] = targets.map(mean_dict)
    env.predict(sample_submission)
    
    count += 1
```
